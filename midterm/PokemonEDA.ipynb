## Question 1

Create as many cells as you need below this one to explore propose, explore and answer your first question.

What kind of correlation exists between Type 1 various Stats (Attack, Defense, Sp. Attak, etc.) and Type 2 and various Stats?

```python
stat_column_names = ["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]
type_column_names = ['Type 1', 'Type 2']
column_names = ['Name', 'Type 1', 'Type 2',"HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]
stat_subset = data.loc[:, column_names]
stat_subset = stat_subset.rename(index=str, columns={"Type 1": "Type_1", "Type 2": "Type_2", "Sp. Atk": "Sp._Atk", "Sp. Def": "Sp._Def"})
```

```python
Chart(stat_subset).mark_bar().encode(
    X('Type_1'),
    Y('HP', aggregate='average')
)
```

```python
Chart(stat_subset).mark_bar().encode(
    X('Type_1'),
    Y('Attack', aggregate='average')
)
```

```python
Chart(stat_subset).mark_bar().encode(
    X('Type_1'),
    Y('Defense', aggregate='average')
)
```
